```python
# Limit operation / limit the records 
for x in mycol.find().limit(3):
    print(x)
```

```python
# Update one record
myquery = {"f_name":"Sunny"}
newvalue = {"$set":{"f_name":"Mohit"}}    # use $set operator , in MongoDB all operator start with $ symbol
mycol.update_one(myquery, newvalue)
```

```python
for x in mycol.find().limit(3):
    print(x)
```

```python
# update one record directly 
mycol.update_one({"last_name":"Savita"},{"$set":{"last_name":"Bhargav"}})
```

```python
for x in mycol.find().limit(3):
    print(x)
```

```python
# performing operator
# finding greater than 
for x in mycol.find({"salary":{"$gt":50000}}):
    print(x)
```

```python
# performing operator
# finding lesser than 
for x in mycol.find({"salary":{"$lt":50000}}):
    print(x)
```
